```python
import pandas as pd
```

```python
data = pd.read_csv("dubai_properties.csv")
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
y = data.price
features = ['latitude', 'longitude', 'size_in_sqft',
       'price_per_sqft', 'no_of_bedrooms', 'no_of_bathrooms']
X = data[features]
```

```python
X.describe()
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=0)
dubaiModel = RandomForestRegressor(random_state=1)
dubaiModel.fit(train_X, train_y)
dubaiPred = dubaiModel.predict(test_X)
```

```python
print(X.head())
print("The prices are: ")
print(dubaiModel.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mae_score = mean_absolute_error(test_y, dubaiPred)
print("Mean Absolute Error: ", mae_score)
```

```python
accuracy = (r2_score(test_y, dubaiPred)*100)
print("Accuracy: {:.2f} %".format(accuracy))
```
